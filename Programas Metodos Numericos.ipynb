```
import matplotlib.pyplot as plt
import numpy as np
import math

if __name__ == "__main__":

  def function(x):
    value = 11567.764*x**2+134821.7633*x**1.85-15
    return value # The main function


  """
    Function of the function.
        Parameters:
        -----------
            x : int | float 
                var
        Returns:
        --------
            value : int | float
                calculated value of the function
  """


  def derivative(x):
    return 23135.53*x + 249420.26*x**0.85 # The derivative of the main function
  """
    Derivative of the function.
        Parameters:
        -----------
            x : int | float 
                var
        Returns:
        --------
            value : int | float
                calculated value of the derivative of the function
  """
    

def newton(function, derivative, x0, tol, max_iter=100):
  x1 = 0
  """
    Calculate and plots the values througth newton-raphson method.
        Parameters:
        -----------
            function : function
                function
            derivative : function
                derivative
            x0: int | float
                x0
            tol: float
                tolerance
            max_iter: int
                number of the maximal iterations
        Returns:
        --------
            x1 : int | float
                calculated value of the function
    """
    

  if abs(x0-x1)<= tol and abs((x0-x1)/x0)<= tol:
    return x0

    print("k\t x0\t\t function(x0)")
    k = 1

    while k <= max_iter:
        x1 = x0 - (function(x0)/derivative(x0))
        print("x%d\t%e\t%e"%(k,x1,function(x1)))

        if abs(x0-x1)<= tol and abs((x0-x1)/x0)<= tol:
            plt.plot(x0, function(x0), 'or')
            return x1

        x0 = x1
        k = k + 1
        plt.plot(x0, function(x0), 'or')

        # Stops the method if it hits the number of maximum iterations
        if k > max_iter:
            print("ERROR: Exceeded max number of iterations")

    return x1  # Returns the value

sqrt = newton(function, derivative, 1.7, 0.0000001)
print("The approximate value of x is: "+str(sqrt))

# Plotting configuration
u = np.arange(0.006, 0.008, 0.0001) # Setting up values for x in the plot
w = 11567.764*u**2+134821.7633*u**1.85-15 # Define the main function again

plt.plot(u, w)
plt.axhline(y=0.0, color='black', linestyle='-')
plt.title('Newton-Raphson Graphics for' + ' y = 11567.764x^2+134821.7633x^1.85-15')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend(['Xn'], loc='upper left')
plt.show()
```

```
from __future__ import division  
import numpy as np  
from numpy import linalg  
 
def jacobi(A,b,x0,tol,N):  
    #preliminares  
    A = array([[6.,1.,1.],
           [1.,9.,-2.],
           [2.,-1.,8.]])
    b = np.array([107., 36., 121.])
    x0 = np.array([1., 1., 1.])
    tol = 0.05
 
    n=np.shape(A)[0]  
    x = np.zeros(n)  
    it = 0  
    #iteracoes  
    while (it < N):  
        it = it+1  
        #iteracao de Jacobi  
        for i in np.arange(n):  
            x[i] = b[i]  
            for j in np.concatenate((np.arange(0,i),np.arange(i+1,n))):  
                x[i] -= A[i,j]*x0[j]  
            x[i] /= A[i,i]  
        #tolerancia  
        if (np.linalg.norm(x-x0,np.inf) < tol):  
            return x  
        #prepara nova iteracao  
        x0 = np.copy(x)  
    raise NameError('num. max. de iteracoes excedido.')
    print(x)
    print(n)
```
